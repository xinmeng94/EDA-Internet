```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_precio_internet = pd.read_csv('Datos/datos de precio de internet.csv')
df_precio_internet
```

```python
df_precio_internet.info()
```

```python
# aunque parece que esta todos los datos completos, vemos que hay muchas columna que presenta 0, es tambien valor desconocido:

df_precio_internet.replace(0., np.nan, inplace=True)
df_precio_internet.isnull().sum()
```

```python
# En este caso podemos ver que hay muchos missing values, usamos el metodo fillna, para quitarlo, ya que 
# normalmente el precio de internet no incrementa mucho en comparación con los años anteriores, segun lo observado con datos.

#df_precio_internet = df_precio_internet.fillna(method="ffill",axis=1)
```

```python
df_precio_internet
```

```python
# dato que hay muchos valores Os, y para el analisis somo necesitamos datos de un año, se realiza la eliminación de los datos no necesarios.

df_precio_internet = df_precio_internet.drop(df_precio_internet.loc[:,"Internet Price, 2010":"Internet Price, 2017"],axis=1)
```

```python
df_precio_internet
```

```python
df_precio_internet.info()
```

```python
# vemos que la columna internet price 2020 no es float con este comando pasamos a tipo float
df_precio_internet['Internet Price, 2020'] = df_precio_internet['Internet Price, 2020'].astype('float64')
```

```python
df_precio_por_pais = df_precio_internet.groupby("Country")[["Internet Price, 2018","Internet Price, 2019","Internet Price, 2020"]].mean()
df_precio_por_pais
```

```python
df_precio_por_pais.info()
```

```python
df_precio_por_pais = df_precio_por_pais.fillna(method="ffill",axis=1)
```

```python
df_precio_por_pais.info()
```

```python
df_precio_por_pais = df_precio_por_pais.drop("Internet Price, 2019",axis=1)
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df_precio_por_pais["Internet Price, 2020"],
             kde=False,
             bins=100);
```

```python
sns.boxplot(df_precio_por_pais["Internet Price, 2020"]);
```

```python
f, (axi1,axi2) = plt.subplots(2, 1, figsize=(10,5), sharex=True);
sns.boxplot(df_precio_por_pais["Internet Price, 2020"],ax=axi1);
sns.distplot(df_precio_por_pais["Internet Price, 2020"],
             kde=False,
             bins=100,ax=axi2);
```

```python
df_precio_por_pais[df_precio_por_pais.isnull().T.any()]
```

```python
p_0 = np.percentile(df_precio_por_pais["Internet Price, 2020"], 0) #Q0 #Min
p_25 = np.percentile(df_precio_por_pais["Internet Price, 2020"], 25) #Q1
p_50 = np.percentile(df_precio_por_pais["Internet Price, 2020"], 50) #Q2 #Mediana 中位数
p_75 = np.percentile(df_precio_por_pais["Internet Price, 2020"], 75) #Q3
p_100 = np.percentile(df_precio_por_pais["Internet Price, 2020"], 100) #Q4 #Max los cualtiles.

print("Valor del percentil 0:", p_0)
print("Valor del percentil 25:", p_25)
print("Valor del percentil 50:", p_50)
print("Valor del percentil 75:", p_75)
print("Valor del percentil 100:", p_100)
print("Valor media:", round(df_precio_por_pais["Internet Price, 2020"].mean(),2) )
```
